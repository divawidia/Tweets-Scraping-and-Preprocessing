Menghilangkan kata @username pada tweet

```python
def remove_pattern(tweet, pattern):
    r = re.findall(pattern, tweet)
    for i in r:
        tweet = re.sub(i, '', tweet)
    return tweet

#menambah tabel remove_user pada dataframe
dataFrame_tweets['remove_user'] = np.vectorize(remove_pattern)(dataFrame_tweets['content'], "@[\w]*")

dataFrame_tweets
```

Menghilangkan link, Angka, Hashtag, dll pada Tweets

```python
def remove(tweet):
    #menghilangkan angka
    tweet = re.sub('[0-9]+', '', tweet)

    tweet = re.sub('_[A-Za-z0-9]+', '', tweet)

    tweet = re.sub('\w+:\/\/\S+', '', tweet)

    tweet = re.sub('www\.[^\s]+', '', tweet)

    tweet = re.sub('www\.[^\s]+', '', tweet)

    tweet = re.sub('https?://[^\s]+', '', tweet)

    tweet = re.sub('http?://[^\s]+', '', tweet)

    tweet = re.sub('\\n', '', tweet)

    tweet = re.sub('\w\.', '', tweet)

    tweet = re.sub('#', '', tweet)

    tweet = re.sub('_', '', tweet)
    
    tweet = re.sub('https?:\/\/.*[\r\n]*,', '', tweet)
    
    return tweet

dataFrame_tweets['remove_link_digit_dll'] = dataFrame_tweets['remove_user'].apply(lambda x: remove(x))
```

Membuat Stemmer dan Stopword

```python
from nltk.corpus import stopwords
stopwords_indonesia = stopwords.words('indonesian')

from Sastrawi.Stemmer.StemmerFactory import StemmerFactory
factory = StemmerFactory()
stemmer = factory.create_stemmer()
```

Mendefinisikan fungsi Tokenisasi dan Stemming Tweets, Menghilangkan Stopword Pada Tweets

```python
from nltk.tokenize import TweetTokenizer

#emot bahagia
emoticons_bahagia = set([
    ':-)', ':)', ';)', ':o)', ':]', ':3', ':c)', ':>', '=]', '8)', '=)', ':}',
    ':^)', ':-D', ':D', '8-D', '8D', 'x-D', 'xD', 'X-D', 'XD', '=-D', '=D',
    '=-3', '=3', ':-))', ":'-)", ":')", ':*', ':^*', '>:P', ':-P', ':P', 'X-P',
    'x-p', 'xp', 'XP', ':-p', ':p', '=p', ':-b', ':b', '>:)', '>;)', '>:-)',
    '<3'
    ])

#emot sedih
emoticons_sedih = set([
    ':L', ':-/', '>:/', ':S', '>:[', ':@', ':-(', ':[', ':-||', '=L', ':<',
    ':-[', ':-<', '=\\', '=/', '>:(', ':(', '>.<', ":'-(", ":'(", ':\\', ':-c',
    ':c', ':{', '>:\\', ';('
    ])    

emoticons = emoticons_bahagia.union(emoticons_sedih)

def bersihkan_tweets(tweet):

    #tokenize tweet
    tokenizer = TweetTokenizer(preserve_case=False, strip_handles=True)
    tweet_tokens = tokenizer.tokenize(tweet)

    tweets_bersih = []
    for word in tweet_tokens:
        if (word not in stopwords_indonesia and word not in emoticons and word not in string.punctuation):
            stem_word = stemmer.stem(word)
            tweets_bersih.append(stem_word)

    return tweets_bersih
```

Mentokenisasi dan Stemming Tweets, Menghilangkan Stopword Pada Tweets dan Membuat kolom baru (tweet_bersih) pada Dataframe

```python
dataFrame_tweets['tweet_bersih'] = dataFrame_tweets['remove_link_digit_dll'].apply(lambda tweets: bersihkan_tweets(tweets))
```

Menghilangkan Tanda Baca

```python
def remove_punct(text):
    text = " ".join([char for char in text if char not in string.punctuation])
    return text
```

```python
dataFrame_tweets['TweetsBersih'] = dataFrame_tweets['tweet_bersih'].apply(lambda tweets: remove_punct(tweets))

dataFrame_tweets
```
